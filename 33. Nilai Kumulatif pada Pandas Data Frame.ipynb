```python
import pandas as pd
```

```python
d = {'pemain': ['budi', 'joni', 'iwan', 'budi', 'budi', 'iwan', 'asep', 'joni'],
    'goal': [2, 1, 3, 1, 1, 2, 2, 3]}

df = pd.DataFrame(d)
df
```

# Menampilkan nilai kumulatif (cumulative)

```python
df['goal'].cumsum().to_frame()
```

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum().to_frame()
df
```

```python
df['jumlah_goal_kumlatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum().to_frame()
df
```

```python
df['cummax'] = df['goal'].cummax()
df
```

```python
df['cummin'] = df['goal'].cummin()
df
```

```python
df['cumprod'] = df['goal'].cumprod()
df
```
