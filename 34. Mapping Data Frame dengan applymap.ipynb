```python
import pandas as pd
```

```python
df = pd.DataFrame({'jenis_kelamin': ['lelaki', 'wanita', 'Lelaki', 'pria', 'perempuan'],
                  'usia': [20, 19, 20, 18, 21],
                  'shift': ['pagi', 'siang', 'Malam', 'malam', 'Pagi']})
df
```

```python
df = df.applymap(lambda x:x.lower() if type(x) == str else x)
df
```

```python
mapping = {'lelaki': 'L',
          'pria': 'L',
          'wanita': 'P',
          'perempuan': 'P',
          'pagi': 1,
          'siang': 2,
          'malam': 3}

df.applymap(mapping.get)
```

```python
df[['jenis_kelamin', 'shift']] = df[['jenis_kelamin', 'shift']].applymap(mapping.get)
df
```
