```python
import pandas as pd
```

```python
df = pd.util.testing.makeMissingDataframe().reset_index()
df.head()
```

```python
df = df.rename(columns={'index':'Z'})
df.head()
```

```python
df_backup = df.copy(deep=True)
```

# Menghapus missing values pada kolom

```python
df = df.dropna(axis='columns')
df.head()
```

# Menghapus missing values pada baris

```python
df = df_backup.copy(deep=True)
df = df.dropna(axis='rows')
df.head()
```

# presentase missing values pada kolom

```python
df = df_backup.copy(deep=True)
df.isna().mean()
```

# Menghapus(drop) setiap kolom yang mengandung missing values berdasarkan thershold

```python
treshold = len(df) * 0.9
df = df.dropna(thresh = treshold, axis='columns')
df.head()
```
