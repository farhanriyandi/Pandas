```python
import pandas as pd
```

# Persiapan Data Frame

```python
data = {'penjual': ['bejo', 'tejo', 'wati', 'bejo', 'cecep', 'tejo', 'wati', 'bejo'],
       'barang': ['monitor', 'keyboard', 'mouse', 'monitor', 'keyboard', 'laptop', 'mouse', 'keyboard']}

df = pd.DataFrame(data)
df
```

```python
df['count_tiap_penjual'] = df.groupby('penjual').cumcount() + 1
df
```

```python
df['count_tiap_barang'] = df.groupby('barang').cumcount() + 1
df
```

```python
df['count_pasangan_kolom'] = df.groupby(['penjual', 'barang']).cumcount() + 1
df
```
